```python
# Daftar kata kunci positif dan negatif
positif_keywords = [
    'msib keren', 'senang', 'bahagia', 'suka', 'bagus', 'memuaskan', 'keren', 'luar biasa',
    'hebat', 'mantap', 'asik', 'cinta', 'damai', 'sempurna', 'menyenangkan', 'puas',
    'terbaik', 'oke', 'salut', 'nyaman', 'positif', 'ceria', 'terima kasih', 'lolos',
    'keterima', 'dapet', 'semoga', 'banget', 'buat', 'temen', 'magang'
]
negatif_keywords = [
    'msib gagal', 'sedih', 'buruk', 'kecewa', 'jelek', 'gagal', 'mengerikan', 'bangsat',
    'marah', 'mengecewakan', 'menyedihkan', 'payah', 'tidak puas', 'parah', 'bodoh',
    'membosankan', 'gagal total', 'jelek banget', 'emosi', 'ga mutu', 'kacau', 'stres',
    'benci', 'tolol', 'basi', 'ga guna', 'gak', 'ga', 'tapi', 'masih', 'udah', 'apa', 'kalo', 'tp'
]
```

```python
# Fungsi pemberian label
def beri_label(text):
    text = str(text).lower()  # Konversi ke huruf kecil
    if any(word in text for word in positif_keywords):
        return 'positif'
    elif any(word in text for word in negatif_keywords):
        return 'negatif'
    else:
        return 'netral'
```

```python
# Tambahkan kolom label ke dataset
df['sentiment'] = df['full_text'].apply(beri_label)
```

```python
# Simpan dataset dengan label baru
labeled_file_path = '/content/drive/MyDrive/A Data Science (Latihan)/Dataset/labeled_data.csv'
df.to_csv(labeled_file_path, index=False)
labeled_file_path
```
